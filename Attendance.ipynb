```python
import csv
from datetime import date

def create_register():
    fname = input("Enter name of the class (ie. XII B): ")
    fmonth = input("Enter the month and year (ie. nov 2023): ")
    f = open(f"{fname}_{fmonth}.csv", 'w', newline='')
    wr = csv.writer(f)
    wr.writerow(["Roll_no", "Name", "Present"])
    print("Register created Successfully!")
    f.close()
    
def add_student():
    try:
        fname = input('Enter name of the class: ')
        fmonth = input("Enter the month and year (ie. nov 2023): ")
        f = open(f"{fname}_{fmonth}.csv", 'a', newline='')
        wr = csv.writer(f)
    except FileNotFoundError:
        print("No such register exists!")
    else:
        n = int(input("Enter number of students to be added: "))
        for i in range(1, n+1, 1):
            rno = int(input(f"Enter roll number of student {i}: "))
            name = input(f"Enter name of student {i}: ")
            wr.writerow([rno, name, list()])
        print("Students added Successfully!")
    finally:
        f.close()
    
def mark_attendance():
    try:
        fname = input('Enter name of the class: ')
        fmonth = input("Enter the month and year (ie. nov 2023): ")
        f = open(f"{fname}_{fmonth}.csv", 'r', newline='')
    except FileNotFoundError:
        print("No such register exists!")
    else:
        datals = list(csv.reader(f))
        print(datals)
        name = input("Enter your name: ")
        with open(f"{fname}_{fmonth}.csv", 'w+', newline='') as fw:
            wr = csv.writer(fw)
            # Linear search
            for i in range(1, len(datals)):
                if datals[i][1].lower() == name.lower():
                    # append attendance date
                    break
            else:
                print('Name of student not found!')
                return -1
            print(datals)
            wr.writerows(datals)
        print("Your Attendance is marked Successfully!")
    finally:
        f.close()

def generate_report():
    try:
        fname = input('Enter name of the class: ')
        fmonth = input("Enter the month and year (ie. nov 2023): ")
        f = open(f"{fname}_{fmonth}.csv", 'r', newline='')
    except FileNotFoundError:
        print("No such register exists!")
    else:
        name = input("Enter name of the student: ")
        datals = list(csv.reader(f))
        for i in range(1, len(datals)):
                if datals[i][1].lower() == name.lower():
                    details = datals[i]
                    break
        else:
            print('Name of student not found!')
            return -1
        print(f"{name}'s Attendance Report: {details}")
    finally:
        f.close()
    
print('''Welcome to Attendance Register!
This is the list of services offered: 
\t1. Creating a new register file for classes according to month and year
\t2. Adding names of student to the register
\t3. Marking the attendance of student
\t4. Generate the attendance report of a Student
\t5. Exit\n''')

choice = int(input("Enter option number: "))
while choice != 5:
    if choice == 1:
        create_register()
    elif choice == 2:
        add_student()
    elif choice == 3:
        mark_attendance()
    elif choice == 4:
        generate_report()
    else:
        print("Enter valid option!")
    print()
    choice = int(input("Enter option number: "))
        
print("Thank You!")
```
